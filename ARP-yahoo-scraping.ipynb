```python
# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, "html.parser")
```

```python
# Find all news items on the page
news_items = soup.find_all("li", class_="js-stream-content")
```

```python
news_items
```

```python
# Initialize an empty list to store the selected news items
selected_news = []
```

```python
# Loop through each news item and extract relevant information
for news_item in news_items:
    headline = news_item.find("h3").text.strip()
    description = news_item.find("p").text.strip()
    date_time_writer_elem = news_item.find("div", class_="C(#959595) Fz(11px) D(ib) Mb(6px)")

    # Check if date_time_writer_elem is not None before extracting information
    if date_time_writer_elem:
        date_time_writer_parts = date_time_writer_elem.find_all("span")
        date_time = date_time_writer_parts[1].text.strip()
        writer = date_time_writer_parts[0].text.strip()

        # Check if "Netflix" is present in the headline
        if "Netflix" in headline:
            selected_news.append({
                "Headline": headline,
                "Description": description,
                "Date and Time": date_time,
                "Writer": writer
            })
```

```python
# Extract numerical values from the "Date and Time" strings
df["Time Value"] = df["Date and Time"].str.extract("(\d+)").astype(int)

# Check if the time is given in hours or days and convert to timedelta
df["Time Delta"] = df.apply(lambda row: timedelta(hours=row["Time Value"]) if "hour" in row["Date and Time"] else timedelta(days=row["Time Value"]), axis=1)

# Get the current date and time when the code was run
current_date_time = datetime(2023, 8, 8, 19, 0)  # Update with the actual date and time

# Calculate the approximate release date based on the time difference
df["Approx Release Date"] = current_date_time - df["Time Delta"]

# Extract and keep only the date from "Date and Time" column
df["Approx Release Date"] = df["Approx Release Date"].dt.date
```
