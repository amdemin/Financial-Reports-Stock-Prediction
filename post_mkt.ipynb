```python
import os
import pandas as pd
import datefinder
from PyPDF2 import PdfReader

def extract_date(pdf_path):
    with open(pdf_path, "rb") as pdf_file:
        reader = PdfReader(pdf_file)
        first_page_text = reader.pages[0].extract_text()

        # Use datefinder to extract dates
        matches = datefinder.find_dates(first_page_text)
        for match in matches:
            return match.strftime('%B %d, %Y')

        return "Date not found"

pdf_folder = "/home/jit/Downloads/arp/ARP-artem/ShareholderLetters"

pdf_info_list = []

for pdf_filename in os.listdir(pdf_folder):
    if pdf_filename.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder, pdf_filename)
        date = extract_date(pdf_path)
        
        if date == "Date not found":
            print(f"Date not found for: {pdf_filename}")
        else:
            pdf_info_list.append({"PDF Name": pdf_filename, "Date": date})

pdf_df = pd.DataFrame(pdf_info_list)
pdf_df
```

```python
# Define the sentiment analysis function
def openai_sentiment_analysis(text):
    # Call the OpenAI API for sentiment analysis
    response = openai.Completion.create(
        engine="text-davinci-003",  # Choose the most suitable engine for sentiment analysis
        prompt=f"Analyze the sentiment of the following text by giving the overall polarity and subjectivity ratings on the strict range (very negative, moderately negative, slightly negative, neutral, slightly positive, moderately positive, very positive) and corresponding score on the strict ramge (-1,1) for the text: {text}" ,
        max_tokens=2000,  # Adjust the number of tokens based on your needs
        temperature=0.0  # Set the temperature to make the response more focused
    )

    # Extract the sentiment analysis result from the API response
    sentiment = response.choices[0].text.strip()

    return sentiment

```

```python
# Test the sentiment analysis function
text_to_analyze = ndf.article_text[0]
sentiment_result = openai_sentiment_analysis(text_to_analyze)
print(sentiment_result)
```

```python
# Add a new column to store sentiment responses
ndf['sentiment_response'] = None

# Loop through the DataFrame and analyze sentiment for each article
for index, row in ndf.iterrows():
    text_to_analyze = row['article_text']
    sentiment_result = openai_sentiment_analysis(text_to_analyze)
    ndf.loc[index, 'sentiment_response'] = sentiment_result

```

```python
print(ndf.sentiment_response.head())
```

```python

# Initialize lists to store extracted values
polarity_values = []
subjectivity_values = []
score_values = []

# Iterate through each row in sentiment_response
for response in ndf['sentiment_response']:
    response_parts = response.split('\n')
    polarity_values.append(response_parts[0].split(': ')[1])
    subjectivity_values.append(response_parts[1].split(': ')[1])
    score_values.append(response_parts[2].split(': ')[1])
```

```python
# Add new columns to the DataFrame
ndf['polarity'] = polarity_values
ndf['subjectivity'] = subjectivity_values
ndf['score'] = score_values

# Drop the original sentiment_response column
ndf.drop('sentiment_response', axis=1, inplace=True)
```
