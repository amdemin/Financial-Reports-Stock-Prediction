```python
# Load the Loughran and McDonald Financial Sentiment Dictionaries
def load_sentiment_dictionary_from_csv(csv_path):
    df = pd.read_csv(csv_path)
    positive_words = set(df[df['Positive'] != 0]['Word'].str.lower())
    negative_words = set(df[df['Negative'] != 0]['Word'].str.lower())
    return positive_words, negative_words

positive_words, negative_words = load_sentiment_dictionary_from_csv("/content/drive/My Drive/ARP/Loughran-McDonald_MasterDictionary_1993-2021.csv")
```

```python
# Step 4: Extract text from the PDF file and preprocess it
def extract_text_from_pdf(pdf_path):
    text = ""
    # Open the PDF file in binary mode
    with open(pdf_path, 'rb') as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)
        # Iterate through all pages and extract text
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text += page.extract_text()
    return text


def preprocess_text(text):
    # Text preprocessing using simple tokenization
    tokens = nltk.word_tokenize(text.lower())
    return tokens
```

```python
# Calculate the polarity score
def calculate_polarity_score(tokens):
    # Count the number of positive and negative words in the text
    positive_count = len(set(tokens).intersection(positive_words))
    negative_count = len(set(tokens).intersection(negative_words))
    # Calculate the polarity score as the difference between positive and negative word counts
    # divided by the total number of words
    polarity_score = (positive_count - negative_count) / len(tokens)
    return polarity_score

def analyze_shareholder_letter(pdf_path):
    # Extract text from the PDF and preprocess it
    text = extract_text_from_pdf(pdf_path)
    tokens = preprocess_text(text)
    # Calculate the polarity score based on the tokens
    polarity_score = calculate_polarity_score(tokens)
    return polarity_score

polarity_score = analyze_shareholder_letter("/content/drive/My Drive/ARP/Final-Q1-23-Shareholder-Letter.pdf")
print("Polarity Score:", polarity_score)
```
